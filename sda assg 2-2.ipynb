```python
import requests
from bs4 import BeautifulSoup
import re

# Specify the URL of the web page
url = "https://macleans.ca/politics/federal-leaders-debate-full-transcript/#section1"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Print the parsed HTML to inspect the structure
    print(soup.prettify())

    # Find the section containing the transcript
    transcript_section = soup.find("div", {"class": "your-identified-class"})
    
    if transcript_section:
        # Extract text from the transcript section
        transcript_text = transcript_section.get_text()

        # Use regular expression to find all words in the text
        words = re.findall(r'\b\w+\b', transcript_text)

        # Print the list of words
        print(words[:10])  # Print the first 10 words as an example
    else:
        print("Transcript section not found.")
else:
    print("Failed to retrieve the web page. Status code:", response.status_code)
```
