```python
mySession1 = requests.session()

response1 = mySession1.get("https://myanimelist.net/anime/50265/Spy_x_Family/reviews")
soup2 = bs(response1.content, 'html.parser')
print(soup2)
```

```python
reviewTable1 = soup2.find("div", {"class": "rightside js-scrollfix-bottom-rel"})
print(reviewTable1)
# brings things like all the text and names of the part of the website you want
```

```python
# recivese the names and texts to create a table
reviewers1=[]
reviewText1=[]

reviewers1 += [r.get_text() for r in reviewTable1.select(".review-element .thumbbody .body .username")]

# retrieve the review text
reviewText1 += [t.get_text() for t in reviewTable1.select(".review-element .thumbbody .body .text")]
```

```python
reviewers12 = []
for sub in reviewers1:
    reviewers12.append(sub.replace("\n", ""))
    
reviewText12 = []
for sub in reviewText1:
    reviewText12.append(sub.replace("\n", ""))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
d2 = {"commenter":reviewers12, "reviews":reviewText12}

# creates a dataframa
df1 = pd.DataFrame(d2) 

df1
#this prints out the table to make it more clear
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
#https://stackoverflow.com/questions/61608057/output-vader-sentiment-scores-in-columns-based-on-dataframe-rows-of-tweets
sia = SentimentIntensityAnalyzer()
df1['compound'] = [sia.polarity_scores(x)['compound'] for x in df1['reviews']]
df1['neg'] = [sia.polarity_scores(x)['neg'] for x in df1['reviews']]
df1['neu'] = [sia.polarity_scores(x)['neu'] for x in df1['reviews']]
df1['pos'] = [sia.polarity_scores(x)['pos'] for x in df1['reviews']]
```
