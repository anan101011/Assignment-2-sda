```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer



# Sample sentence for sentiment analysis
a = "The anime is good"

# Initialize the SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Calculate sentiment scores for the sample sentence
sentiment_scores = sid.polarity_scores(a)

# Add sentiment scores as new columns to the DataFrame
df1['SentimentCompound'] = df1['Text'].apply(lambda x: sid.polarity_scores(x)['compound'])
df1['SentimentNegative'] = df1['Text'].apply(lambda x: sid.polarity_scores(x)['neg'])
df1['SentimentNeutral'] = df1['Text'].apply(lambda x: sid.polarity_scores(x)['neu'])
df1['SentimentPositive'] = df1['Text'].apply(lambda x: sid.polarity_scores(x)['pos'])

# Display the updated DataFrame
print(df1)
```
